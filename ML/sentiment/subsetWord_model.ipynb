```python
import graphlab
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
products = graphlab.SFrame('amazon_baby.gl/')
```

```python
# use only subset of words to train model

selected_words = ['awesome', 'great', 'fantastic', 'amazing', \
                  'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
products['word_counts'] = graphlab.text_analytics.count_words(products['review'])
products.head()
```

```python
products['word_counts'][0]
```

```python
def awesome_count(word_count):
    # input is per row basis
    if 'awesome' in word_count:
        num = word_count['awesome']
    else:
        num = 0
    return num
```

```python
products["awesome"] = products['word_counts'].apply(awesome_count)
```

```python
products.head()
```

```python
def great_count(word_count):
    # input is per row basis
    if 'great' in word_count:
        num = word_count['great']
    else:
        num = 0
    return num

def fantastic_count(word_count):
    # input is per row basis
    if 'fantastic' in word_count:
        num = word_count['fantastic']
    else:
        num = 0
    return num

def amazing_count(word_count):
    # input is per row basis
    if 'amazing' in word_count:
        num = word_count['amazing']
    else:
        num = 0
    return num

def love_count(word_count):
    # input is per row basis
    if 'love' in word_count:
        num = word_count['love']
    else:
        num = 0
    return num

def horrible_count(word_count):
    # input is per row basis
    if 'horrible' in word_count:
        num = word_count['horrible']
    else:
        num = 0
    return num

def bad_count(word_count):
    # input is per row basis
    if 'bad' in word_count:
        num = word_count['bad']
    else:
        num = 0
    return num

def terrible_count(word_count):
    # input is per row basis
    if 'terrible' in word_count:
        num = word_count['terrible']
    else:
        num = 0
    return num

def awful_count(word_count):
    # input is per row basis
    if 'awful' in word_count:
        num = word_count['awful']
    else:
        num = 0
    return num

def wow_count(word_count):
    # input is per row basis
    if 'wow' in word_count:
        num = word_count['wow']
    else:
        num = 0
    return num

def hate_count(word_count):
    # input is per row basis
    if 'hate' in word_count:
        num = word_count['hate']
    else:
        num = 0
    return num
```

```python
products['hate'] = products['word_counts'].apply(hate_count)
products['wow'] = products['word_counts'].apply(wow_count)
products['love'] = products['word_counts'].apply(love_count)
products['amazing'] = products['word_counts'].apply(amazing_count)
products['horrible'] = products['word_counts'].apply(horrible_count)
products['terrible'] = products['word_counts'].apply(terrible_count)
products['awful'] = products['word_counts'].apply(awful_count)
products['great'] = products['word_counts'].apply(great_count)
products['love'] = products['word_counts'].apply(love_count)
products['fantastic'] = products['word_counts'].apply(fantastic_count)
products['bad'] = products['word_counts'].apply(bad_count)
```

```python
products.head()
```

```python
for _, word in enumerate(selected_words):
    print word
    print products[word].sum()
```

```python
products = products[products['rating'] != 3]    # remove 3* reviews
products['sentiment'] = products['rating'] >= 4 
```

```python
train_data, test_data = products.random_split(0.8, seed=0)
features = selected_words
```

```python
subsetWords_model = graphlab.logistic_classifier.create(train_data, target='sentiment',
                                    features=features, validation_set=test_data)
```

```python
subsetWords_model.evaluate(test_data)
```

```python
subsetWords_model.evaluate(test_data,  metric='roc_curve')
```

```python
subsetWords_model.show(view='Evaluation')
```

```python
# examine weights of the learned classifier
subsetWords_model['coefficients'].sort('value', ascending=False)
```

```python
products['predicted_sentiment'] = subsetWords_model.predict(products, output_type='probability')
```
